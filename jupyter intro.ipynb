```python
from scipy import ndimage
im = ndimage.imread('real.jpg', mode='L')
plt.imshow(im)
plt.show()
im.shape
```

```python
# get the car
plt.imshow(im[350:550, 1000:1300])
```

```python
plt.hist(im[350:550, 1000:1300].flatten())
```

```python
subimage = im[350:550, 1000:1300].copy()
subimage[subimage < 50] = 0
plt.figure()
plt.imshow(subimage)
subimage[subimage > 75] = 0
plt.figure()
plt.imshow(subimage)

```
